```python
with open('smartprix_page.html','r',encoding='utf-8') as f:
    html = f.read()
```

```python
from bs4 import BeautifulSoup
import numpy as np
import pandas as pd
```

```python
soup = BeautifulSoup(html,'lxml')
```

```python
containers = soup.find_all('div',{'class':'sm-product has-tag has-features has-actions'})
```

```python
# Make features name
```

```python
Product_Name = []
prices = []
Discount_Percentage = []
processor = []
Processor_Details=[]
RAM=[]
Storage=[]
Graphics=[]
Display_Size=[]
Operating_System=[]
Warranty_Period=[]
Space_Score=[]
```

```python
for i in containers:
    try:
        Product_Name.append(i.find('h2').text)
    except:
        Product_Name.apped(np.nan)
    try:    
        prices.append(i.find('span',{'class':'price'}).text)
    except:
        prices.append(np.nan)
    try:    
        Discount_Percentage.append(i.find('span',{'class':'sm-pdrop'}).text)
    except:
        Discount_Percentage.append(np.nan)


    #extract all feature of leptop 

    ul_element = soup.find('ul', class_='sm-feat specs')
    # Extract the text from each <li> and join with a newline
    list_items = ul_element.find_all('li')
    separator = '\n'
    result = separator.join(item.get_text(strip=True) for item in list_items)

    lines = result.splitlines()
    try:
        processor.append(lines[0])
    except:
        processor.append(np.nan)
    try:
        Processor_Details.append(lines[1])
    except:
        Processor_Details.append(np.nan)
    try:    
        RAM.append(lines[2])
    except:
        RAM.append(np.nan)
    try:
        Storage.append(lines[3])
    except:
        Storage.append(np.nan)
    try:    
        Graphics.append(lines[4])
    except:
        Graphics.append(np.nan)
    try:
        Display_Size.append(lines[5])
    except:
        Display_Size.append(np.nan)
    try:
        Operating_System.append(lines[6])
    except:
        Operating_System.append(np.nan)
    try:    
        Warranty_Period.append(lines[7])
    except:
        Warranty_Period.append(np.nan)
    
#     for srore rank
    
    try:
        Space_Score.append(i.find('div',{'class':'score rank-3-bg'}).find('b').text)
    except:
        Space_Score.append(np.nan)
```

```python
# creating table here
```

```python
df = pd.DataFrame({
'Product_Name':Product_Name,
'prices':prices,
'Discount_Percentage':Discount_Percentage ,
'processor':processor,
'Processor_Details':Processor_Details,
'RAM':RAM,
'Storage':Storage,
'Graphics':Graphics,
'Display_Size':Display_Size,
'Operating_System':Operating_System,
'Warranty_Period':Warranty_Period,
'Space_Score':Space_Score,
})
```
